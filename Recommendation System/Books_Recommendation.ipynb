```python
books=books.drop_duplicates(subset=['ID','Book.Title'])
books.head()
```

```python
books.reset_index(inplace=True,drop=True)
books.head()
```

__Creating Pivot Table of our data__

```python
new_book=books.pivot_table(index='ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
```

```python
new_book.index = books.ID.unique()
```

```python
new_book
```

```python
new_book.fillna(0, inplace=True)
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
Book = 1 - pairwise_distances(new_book.values,metric='cosine')
```

```python
Book
```

```python
#Store the results in a dataframe
user_Title = pd.DataFrame(Book)
user_Title
```

```python
#Set the index and column names to user ids 
user_Title.index = books.ID.unique()
user_Title.columns = books.ID.unique()
user_Title
```

```python
np.fill_diagonal(Book, 0)
user_Title.iloc[0:5, 0:5]
```
